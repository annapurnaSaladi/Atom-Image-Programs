```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure()
dark = np.fromfile('absorption_img01/dark.bin', dtype='float64', sep="")
dark = dark.reshape([140, 512])
plt.imshow(dark)
```

```python
light = np.fromfile('absorption_img01/reference.bin', dtype='float64', sep="")
light = light.reshape([140, 512])
plt.imshow(light)
```

```python
atoms = np.fromfile('absorption_img01/atoms.bin', dtype='float64', sep="")
atoms = atoms.reshape([140, 512])
plt.imshow(atoms)
```

```python
OD = np.log((light-dark)/(atoms-dark))
OD = OD[67:92, 149:225]

plt.imshow(OD)

print(OD.min())
print(OD.max())

lam = 460.8*(10**-9)
sig = (3*(lam**2))/(2*np.pi)
```

```python
n = (1/sig)*np.sum(OD*(((16*(10**(-6)))/2)**2))
print(n)
```
